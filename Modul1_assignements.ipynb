Write a function to check if the year number is a leap year

```python
year =int(input("Enter the year"))
def leap_year(year):
  if (year%4==0 and year%100!=0) or year %400==0:
    print(f"{year} is a leap year")
  else:
    print(f"{year} is not a leap year")
leap_year(year)
```

Write a function that takes 2 arrays and prints the members of the first
array that are present in the second array. Hint: use Membership comprehension

```python

def array_existing(ls1 , ls2):
  common_element=[]
  for i in ls1:
    if i in ls2:
      common_element.append(i)
  return common_element
def array_existing_comprehesion(ls1 , ls2):
  common_element= [i for i in ls1 if i in ls2]
  return common_element

ls1 =[23,12,24,1123,45,32,46]
ls2 =[12,32,45, 56]

print(array_existing(ls1,ls2))
print(array_existing_comprehesion(ls1,ls2))
```
