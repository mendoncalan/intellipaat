```python
# Create a list of the first 50 natural numbers
# Count the even numbers from the list
# Return the sum of the list
# Reverse the list

# Time till 8:33 pm
```

```python
def create_list():
  ls =[]
  count=0
  sum =0
  for i in range(1,51):
    if i %2==0:
      count+=1
    sum+=i
    ls.append(i)
  print(f"Count of even Number: {count}")
  print(f"List {ls}")
  print(f"Sum: {sum}")
  print(f"Reverse :{ls[::-1]}") # start index: stop index : step size
create_list()
```

```python
Number1 = int(input("Enter the Number1: "))
Number2 = int(input("Enter the Number2: "))


def list_operations():
    numbers = list(range(Number1, Number2))
    print("List:", numbers)


    Even = 0
    for num in numbers:
        if num % 2 == 0:
            Even += 1
    print("Count of even numbers:", Even)


    total_sum = sum(numbers)
    print("Sum of the list:", total_sum)


    numbers.reverse()
    print("Reversed list:", numbers)


list_operations()
print("Code is Complted")
```

```python
# Write the code to count the numbers of vowel in a string.....

# Time till 8:50 pm
```

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
input_string = input("Enter a string: ")
vowel_count = count_vowels(input_string)
print(f" vowels  is: {vowel_count}")
```

```python
# Write a function to check whether a number is prime or not

# Time till 9:04 pm
```

```python
number = int(input("Enter a number: "))
def check_prime(number):
    if number <= 1:
        print("Not a Prime Number")
        return


    for i in range(2, number):
        if number % i == 0:
            print("Not a Prime Number")
            return


    print(number, "is a Prime Number")


check_prime(number)
```

```python
# Write the code to find the sum of digits of a number

# 479 = 20

# Time till 9:30
```

```python
number = int(input("Enter a number: "))
def Sum(number):
    total = 0
    for digit in str(number):
        total = total + int(digit)
    print("Sum of digits:", total)


Sum(number)
print("Code is Completed.")
```

```python
#  Write to code to find the Count,  how many times a given element appears in a list

# Time till 10:03 pm
```

```python
numbers = eval(input("Enter list elements (comma separated): "))
target = int(input("Enter element to count: "))


def count_element(list, element):
    count = 0
    for item in list:
        if item == element:
            count += 1
    print("Element appears", count, "times")


count_element(numbers, target)
print("Code is Completed")
```

```python
# Write the code to find the factorial of a number

# 5! = 5*4*3*2*1 = 120

# Time till 10:37 pm
```

```python
def factorial(num):
  fact=1
  for i in range(1,num+1):
    fact=fact*i
  return fact
factorial(int(input("Enter number")))
```
