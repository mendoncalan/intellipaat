write a function to check weather a number is number prime or composite

```python
def check_prime_numb(num):
  prime =1
  for i in range(2,int(num*0.5)):
    if num % i ==0:
      prime =0
      break

  if prime==1:
    print(f"{num} is prime number")
  else:
    print(f"{num} is composite number")
check_prime_numb(int(input("Enter number")))
```

```python
n = int(input("Enter a number: "))
def prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            print("Not a prime number")
    print("Prime number")

prime(n)
```
