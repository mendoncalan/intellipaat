**String Indexing** :Strings are ordered and indexed (starting at 0).


```python
#Indexing in python
#Index: position number of each letter
name ="Python"
#Python index always starts from 0
print(name[0])
print(name.index('P'))
```

```python
name ='pappu'
#Python index always returns the first occurence
print(name.index('p'))
```

```python
#Negative index in python starts from -1
#Python
# -6 -5 -4 -3 -2 -1
name = 'Python'
print(name[-1])
```

```python
# Take a string from user end
# Print first character
# print last character (using negative index )
# print middle character
# print index of the first occurence of a
# print seconed last character

String = input("Enter string ")
print("First charater ",String[0])
print("Last character ",String[-1])
print("Middle character ",String[len(String)//2]  )
print ("Index of the first occurence of  a ",String.index(('a')))
print ("Second last index of String ", String[-2] )
```

**String Slicing Syntax: string[start:end:step]**

```python
word = "Python"
print(word[0:4])   # Pyth
print(word[:3])    # Pyt
print(word[3:])    # hon
print(word[::2])   # Pto
print(word[::-1])  # nohtyP (reverse)
```

```python
#String Immutability (IMPORTANT):Strings cannot be changed after creation.
word = "Python"
word[0] = "J"   # Error
```

```python
word = "Python"
word = "J" + word[1:]
print(word)  # Jython
```

**String Methods (Very Important)**

```python
#Case Methods
text = "hello world"

print(text.upper())     # HELLO WORLD
print(text.lower())     # hello world
print(text.title())     # Hello World
print(text.capitalize())# Hello world
```

```python
#Remove Spaces
text = "  hello  "

print(text.strip())   # remove both sides
print(text.lstrip())  # left only
print(text.rstrip())  # right only
```

```python
#Search in String
text = "Python Programming"

print(text.find("Pro"))   # 7
print(text.index("Python"))  # 0
print("Python" in text)   # True
```

```python
#Replace
text = "I like Java"

print(text.replace("Java", "Python"))
```

```python
#Split & Join
text = "apple,banana,orange"

items = text.split(",")
print(items)

print("-".join(items))
```

```python
#Length of String
text = "Python"
print(len(text))   # 6
```

**String Formatting (Very Important)**

```python

#Old Style
name = "Alice"
age = 25
print("My name is %s and I am %d years old" % (name, age))


# format() Method
print("My name is {} and I am {}".format(name, age))

#f-Strings (Best Method)
print(f"My name is {name} and I am {age}")
```

**String Checks**

```python
text = "Python123"

print(text.isalpha())   # False
print(text.isdigit())   # False
print(text.isalnum())   # True
print(text.isspace())   # False
```
