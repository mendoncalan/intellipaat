write a function to check weather a number is number prime or composite

```python
def check_prime_numb(num):
  prime =1
  for i in range(2,int(num*0.5)):
    if num % i ==0:
      prime =0
      break

  if prime==1:
    print(f"{num} is prime number")
  else:
    print(f"{num} is composite number")
check_prime_numb(int(input("Enter number")))
```

```python
n = int(input("Enter a number: "))
def prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            print("Not a prime number")
    print("Prime number")

prime(n)
```

```python
# write a code to remove a give element all occurence from a list

def create_list(len_ls):
  ls=[]
  for i in range(0,len_ls):
    ls.append(int(input("Enter number")))
  print(f"List: {ls}")
  return ls
def remove_element(ls,num):
  l=len(ls)
  while l >0:
    if ls[l-1]==num:
      ls.pop(l-1)
    l-=1
  print(f"List: {ls}")
len_ls =int(input("Enter lenght of list"))
ls=create_list(len_ls)
num=int(input("Enter number"))
remove_element(ls,num)
```

```python
# wirte the code to find the factorial of a number

def factorial(num):
  fact=1
  for i in range(1,num+1):
    fact=fact*i
  return fact
factorial(int(input("Enter number")))
```
