Operators in python
Special symbol that help you perform operation on values(operands)
1.  Arithmatic operator : +/*-%
2.  Exponential Operator
3.  comparison operator
4.  logical Operator
5.  Assignment operator


**Arithmatic operator**:Used for mathematical calculations.
| Operator | Meaning             | Example  | Result |
| -------- | ------------------- | -------- | ------ |
| `+`      | Addition            | `5 + 3`  | `8`    |
| `-`      | Subtraction         | `5 - 3`  | `2`    |
| `*`      | Multiplication      | `5 * 3`  | `15`   |
| `/`      | Division            | `5 / 2`  | `2.5`  |
| `//`     | Floor Division      | `5 // 2` | `2`    |
| `%`      | Modulus (remainder) | `5 % 2`  | `1`    |
| `**`     | Power               | `2 ** 3` | `8`    |


**Comparison Operators**:Used to compare values. Result is True or False.

| Operator | Meaning |
| -------- | ------------------- |
|==	| Equal to|
|!=	|  Not equal to|
|>	|    Greater than|
|<	|Less than|
|>=	|Greater than or equal|
|<=	|Less than or equal|

**Logical Operators**:Used to combine conditions.
| Operator | Meaning |
| -------- | ------------------- |
|and|	True if both conditions are True|
|or|	True if at least one is True|
|not|	Reverses the result|


**Bitwise Operators**:Used for binary operations (advanced).
| Operator | Meaning |
| -------- | ------------------- |
|&|	AND|
|`|	`|
|^|	XOR|
|~|	NOT|
|<<|	Left shift|
|>>|	Right shift|

**Assignment Operators**: Used to assign values to variables.

| Operator | Example  | Same As     |
| -------- | -------- | ----------- |
| `=`      | `x = 5`  | Assign 5    |
| `+=`     | `x += 3` | `x = x + 3` |
| `-=`     | `x -= 3` | `x = x - 3` |
| `*=`     | `x *= 2` | `x = x * 2` |
| `/=`     | `x /= 2` | `x = x / 2` |


**Membership Operators** : Used to test if value exists in a sequence.


| Operator | Meaning              |
| -------- | -------------------- |
| `in`     | Value exists         |
| `not in` | Value does not exist |

**Identity Operators**:Check if two variables refer to same object in memory.
| Operator | Meaning         |
| -------- | --------------- |
| `is`     | Same object     |
| `is not` | Not same object |

What is a python Token?
In Python, a token is the smallest individual unit of a program that has meaning to the interpreter. You can think of tokens as the "words" and "punctuation" that make up the "sentences" of your code


Token Type,Description,Examples
Keywords,"Reserved words with a fixed, predefined meaning.","if, while, def, return, True"
Identifiers,"Names you give to variables, functions, or classes.","my_var, calculate_sum, User"
Literals,Fixed data values (raw data).,"42, 3.14, ""Hello"", None"
Operators,Symbols used for computations or logic.,"+, -, *, /, ==, and"
Delimiters,Symbols used for grouping and syntax structure.,"( ), [ ], { }, :, ,, ="

#Set: Unordered  

```python
s1 = {'payal','Aman', True, 34.67}
print(type(s1))
# adding element in Set. it will add element any where in the set
s1.add('Alan')
print(s1)
# Remove basically remvoes a specific element by value from the set. if the value is not present remove will give us an error.
s1.remove(34.67)
print(s1)
# difference between discard and remove
# discard: it removed the specific element from the set and if it is not available it doesnt give any error.
s1.discard(34.67)
print(s1)

#Membership Test
numbers = {1, 2, 3}

print(2 in numbers)   # True

#Set with Mixed Data
#Tuple allowed
#List not allowed
my_set = {1, "a", (2, 3)}
print(my_set)
# my_set = {[1,2]}# error TypeError: unhashable type: 'list'

my_set = {1, True, 0, False}
print(my_set)#{1, 0}


#Frozen Set (Advanced):If you want an immutable set:
fs = frozenset([1, 2, 3])

#Set Comprehension
#synatx
 #{expression for item in iterable}"
#{expression for item in iterable if condition}

squares = {x*x for x in range(5)}
print(squares)

#Remove Duplicates Automatically
numbers = [1, 2, 2, 3, 3, 4]

unique_squares = {x*x for x in numbers}
print(unique_squares)

#With Condition
evens = {x for x in range(10) if x % 2 == 0}
print(evens)

#Convert Strings to Uppercase
words = ["apple", "banana", "apple"]

upper_words = {word.upper() for word in words}
print(upper_words)

#Extract Unique Characters from String
text = "programming"

unique_chars = {char for char in text}
print(unique_chars)

#Nested Loop in Set Comprehension
pairs = {(x, y) for x in range(3) for y in range(2)}
print(pairs)
```

```python
#Simple Dictionary Example
person = {
    "name": "Alan",
    "age": 35,
    "city": "Mumbai"
}

print(person["name"])
#Adding & Updating Values
person["occupation"] = "Engineer"
person["age"] = 36
print(person)

#Looping Through Dictionary
  #Loop through keys
for key in person:
    print(key)

  #Loop through values
for value in person.values():
    print(value)
  #Loop through key-value pairs
for key, value in person.items():
    print(key, ":", value)
```

```python
#Nested Dictionary
students = {
    "s1": {"name": "Alan", "marks": 85},
    "s2": {"name": "John", "marks": 90}
}

print(students["s1"]["marks"])
#Dictionary Inside List
employees = [
    {"id": 1, "name": "Alan"},
    {"id": 2, "name": "John"}
]

print(employees[0]["name"])
```

```python
#Counting Frequency
text = "apple banana apple mango banana apple"

words = text.split()
count = {}

for word in words:
    count[word] = count.get(word, 0) + 1

print(count)
```

```python
#Using Dictionary Comprehension
numbers = [1, 2, 3, 4]

square_dict = {x: x*x for x in numbers}
print(square_dict)

#Merge Two Dictionaries
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}
#The ** operator is called dictionary unpacking.it will convert {"a": 1, "b": 2} to "a": 1, "b": 2
merged = {**d1, **d2}
print(merged)
#Check If Key Exists
if "name" in person:
    print("Key exists")


#Sorting Dictionary by Key
data = {"b": 2, "a": 1, "c": 3}

sorted_dict = dict(sorted(data.items()))
print(sorted_dict)

#Sorting by Value
sorted_by_value = dict(sorted(data.items(), key=lambda x: x[1]))
print(sorted_by_value)
#key=lambda x: x[1]
#This means:
#Take each tuple (key, value)
#Sort based on index 1 → the value
```

**String**

> A sequence of characters enclosed in quotes.
Important Properties


1.  Ordered
2.  Immutable (cannot modify directly)
3.  Indexed
4.  Iterable
5.  Allows duplicates


```python
#Indexing
text = "Python"

print(text[0])   # P
print(text[-1])  # n

#Slicing
text = "Python"

print(text[0:4])   # Pyth
print(text[:3])    # Pyt
print(text[3:])    # hon
print(text[::-1])  # nohtyP (reverse)

#Case Conversion Functions
text = "python"

print(text.upper())      # PYTHON
print(text.lower())      # python
print(text.capitalize()) # Python
print(text.title())      # Python
print(text.swapcase())   # PYTHON

#Search & Find Functions

text = "hello world"

print(text.find("world"))   # 6
print(text.index("world"))  # 6
print(text.count("l"))      # 3

#Difference:
#find() → returns -1 if not found
#index() → raises error if not found

#Check Functions

text = "Python123"

print(text.isalpha())   # False
print(text.isdigit())   # False
print(text.isalnum())   # True
print(text.islower())   # False
print(text.isupper())   # False
print(text.isspace())   # False

#Replace Function
text = "I love Java"

new_text = text.replace("Java", "Python")
print(new_text)
#Strings are immutable → new string created.


#Split & Join
text = "apple,banana,mango"

print(text.split(","))  #['apple', 'banana', 'mango']

#Join
words = ["apple", "banana", "mango"]

result = ",".join(words)
print(result) #apple,banana,mango

#Strip Functions
text = "   hello   "

print(text.strip())     # remove both sides
print(text.lstrip())    # left only
print(text.rstrip())    # right only


#Startswith & Endswith
filename = "error.log"

print(filename.startswith("error"))  # True
print(filename.endswith(".log"))     # True

#Formatting Strings
#old way
name = "Alan"
print("Hello {}".format(name))
#New way
name = "Alan"
print(f"Hello {name}")

#String Immutability
text = "hello"
#text[0] = "H"   #  Error
# right way to change the first
text = "H" + text[1:]


#String Concatenation
a = "Hello"
b = "World"

print(a + " " + b)

#Membership Check
text = "python"

print("py" in text)   # True

#Iterating Over String
for char in "Python":
    print(char)


#Reverse words in a sentence:
text = "I love Python"

words = text.split()
reversed_sentence = " ".join(words[::-1])

print(reversed_sentence)

#print every 2nd lette
even_text= text[::2]
print(even_text)

# Remove *** from string
text ="****hello****"
print(text.strip('*'))
print()
```
